```python
# Resposta questão 5 A: 
import numpy as np
import matplotlib.pyplot as plt

recs =df['Sexo'].value_counts().tolist()

labels = ['Sexo']
homens = [recs[0]]
mulheres = [recs[1]]

x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, homens, width, label='Homens')
rects2 = ax.bar(x + width/2, mulheres, width, label='Mulheres')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Reclamacoes')
ax.set_title('Reclamacoes')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()


def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(7, 0.1),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')


autolabel(rects1)
autolabel(rects2)

fig.tight_layout()

plt.show()
```

```python
# Resposta questao 5 B

import matplotlib.pyplot as plt
import numpy as np

reclamacao_uf = df['UF'].value_counts().keys().tolist()

pos = np.arange(len(reclamacao_uf))

reclamacao = df['UF'].value_counts().tolist()

plt.bar(pos,reclamacao,color='blue',edgecolor='black')

plt.xticks(pos, reclamacao_uf)

plt.xlabel('UF', fontsize=16)

plt.ylabel('Reclamacoes', fontsize=16)

plt.title('Reclamacoes por UF',fontsize=20)

plt.show()
```

```python
# Resposta 6 A
#Relacao entre a faixa etaria e a quantidade de reclamacoes 

t = df['Faixa Etária'].value_counts().keys().tolist()
t = [t[6], t[1], t[0], t[2], t[3], t[4], t[5]]

t_r = df['Faixa Etária'].value_counts().tolist()
t_r = [t_r[6], t_r[1], t_r[0], t_r[2], t_r[3], t_r[4], t_r[5]]

fig, axs = plt.subplots()

axs = plt.gca()
axs.set_ylim([0,20000])

plt.plot(t,t_r,'bo-')

for x,y in zip(t,t_r):

    label = y

    plt.annotate(label, # this is the text
                 (x,y), # this is the point to label
                 textcoords="offset points", # how to position the text
                 xytext=(0,10), # distance from text to points (x,y)
                 ha='center') # horizontal alignment can be left, right or center

axs.plot(t,t_r)
plt.xticks(rotation=90)
fig.suptitle('Faixa Etaria e Quantidade de Reclamacoes')
```
